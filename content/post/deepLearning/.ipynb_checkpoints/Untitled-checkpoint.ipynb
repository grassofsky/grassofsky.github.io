# 二元线性回归分析

二元，有两组数据；线性，数据($a_1, a_2, a_3, ..., a_n$)之间的关系，可以用公式$a_1b_1+a_2b_2+...+a_nb_n+b_0=0$进行表示；回归，建立数据之间的关系就叫做回归，对于二元数据，公式表示即为$Y=f(X)$，其中X为自变量，Y为因变量或响应变量。

## 问题

下表中给出了某地区8名儿童年龄（岁）和尿肌酐（mmol/24h）。

| 编号 | 1 | 2 | 3 | 4 | 5 | 6 | 7 | 8 |
| ---- |--|---|----|---|---|---|--|---|
| 年龄X |13 | 11 | 9 | 6 | 8 | 10 | 12 | 7 |
| 尿肌酐含量Y | 3.54 | 3.01 | 3.09 | 2.48 | 2.56 | 3.36 | 3.18 | 2.65 |

问：该地区儿童的年龄和尿肌酐含量之间是否有某种关系？

## 数据展示

为了直观的观察数据之间的关系，我散点图进行绘制


```python
# python imports
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
import matplotlib as mpl
zhfont = mpl.font_manager.FontProperties(fname='/usr/share/fonts/opentype/wqy/wqy-microhei.tcc')
```

```python
# Plot relation between X and Y
X = [13, 11, 9, 6, 8, 10, 12, 7]
Y = [3.54, 3.01, 3.09, 2.48, 2.56, 3.36, 3.18, 2.65]
plt.plot(X,Y,'x')
plt.xlabel('年龄（岁）', fontsize=15, fontproperties=zhfont)
plt.ylabel('尿肌酐（mmol/24h)', fontsize=15, fontproperties=zhfont)
plt.title(u'年龄与尿肌酐之间的关系', fontproperties=zhfont)
plt.grid()
```
